```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
from scipy import stats 
from sklearn.datasets import load_boston

%matplotlib inline 
```

```python
boston=load_boston()
print(boston.keys())
```

```python
df=pd.DataFrame(boston['data'],columns=boston.feature_names)
df.head()
```

```python
df['PRICE']=boston['target']
```

```python
print("Boston 데이터 세트 크기: ",df.shape)
df.head()
```

```python
# 데이터 값의 정보에 대해서 알아봅시다
df.info()
```

null 값은 존재하지 않으며 모든 dtype들이 float의 형태를 띄고 있음을 우리는 알 수 있습니다.

각 칼럼이 회귀에 미치는 영향을 시각화해서 알아보도록 하겠습니다.

```python
fig,axs=plt.subplots(figsize=[16,8],ncols=4,nrows=2)
lm_features=["RM","ZN","INDUS","NOX","AGE","RAD","PTRATIO","LSTAT"]
for i, feature in enumerate(lm_features):
    row=int(i/4)
    col=i%4
    # 시본의 regplot을 이용해 산점도와 선형회귀 직선을 함께표현 
    sns.regplot(x=feature,y='PRICE',data=df,ax=axs[row][col])
```

다른 칼럼들보다 RM,은 양방향의 선형수가 가장크며 LSTAT는 음방향의 선형성이 가장 큽니다. 즉 LSTAT이 적을수록 PRICE가 증가하는 모습을 보여줍니다. 

이제 LinearRegression class를 이용해 보스턴 주택 가격의 회귀 모델을 만들겠습니다. train_test_split()을 이용해 학습과 테스트 데이터 세트를 분리해 학습과 예측을 수행합니다.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

y_target=df['PRICE']
x_data=df.drop(['PRICE'],axis=1)

X_train,X_test,y_train,y_test = train_test_split(x_data,y_target,test_size=0.3,random_state=156)

#선형 회귀로 ols학습/예측/평가
lr=LinearRegression()
model=lr.fit(X_train,y_train)
y_preds=model.predict(X_test)
rmse=np.sqrt(mean_squared_error(y_preds,y_test))

print("MSE : {0:.3f}, RMSE : {1:.3f}".format(mean_squared_error(y_preds,y_test),rmse))
print("Variance Score: {:.3f}".format(r2_score(y_test,y_preds)))
```

```python
print("회귀 계수값:",model.coef_)
print("절편값:",model.intercept_)
```

```python
#회귀계수를 큰 값순으로 정렬하기 위해 Series 생성 index칼람명에 유의 
coeff=pd.Series(data=np.round(model.coef_,1),index=x_data.columns)
coeff.sort_values(ascending=False)
```

RM이 양의 값으로 회귀 계수가 가장 크며, nox피처의 회귀 계수 - 값이 너무 커 보입니다. 차츰 최적화를 수행하면서 피처 coefficients의 변화도 같이 살펴보겠습니다.

이번에는 5개의 폴드세트에서 cross_val_score()를 이용해 교차 검증으로 MSE와 RMSE를 측정해보겠습니다. 싸이킷런은 cross_val_score()를 이용할 텐데 RMSE를 제공하지 않으므로 MSE 수치결과를 ROOT로 바꿔줘야합니다 cross_val_score()의 인자로 scoring="neg_mean_squared_error"를 지정하면 반환되는 값은 음수값입니다. 사이킷런의 지표평가가 높은 값일 수록 좋은 모델인 데 반해 일반적으로 회귀는 mse 값이 낮을수록 좋은 모델이기에 이런식으로 default 지정을 해 놓은 것입니다.

```python
from sklearn.model_selection import cross_val_score

y_target=df['PRICE']
X_data=df.drop(['PRICE'],axis=1)

#cross_val_score()로 5폴드세트로 mse 구현후 이를 기반으로 다시 rmse구함 
neg_mse_scores=cross_val_score(lr,X_data,y_target,scoring="neg_mean_squared_error",cv=5)
rmse_scores=np.sqrt(-1*neg_mse_scores)
avg_rmse=np.mean(rmse_scores)

#cross_val_score(scoring="neg_mean_squared_error")
print("5 folds의 개별 neg_mse_scores:",np.round(neg_mse_scores,2))
print("5 folds 개별 rmse scores:",np.round(rmse_scores,2))
print("평균 rmse:{:.3f}".format(avg_rmse))
```

```python
# 릿지 회귀 
from sklearn.linear_model import Ridge

ridge=Ridge(alpha=10)
#cross_val_score()로 5폴드세트로 mse 구현후 이를 기반으로 다시 rmse구함 
neg_mse_scores=cross_val_score(ridge,X_data,y_target,scoring="neg_mean_squared_error",cv=5)
rmse_scores=np.sqrt(-1*neg_mse_scores)
avg_rmse=np.mean(rmse_scores)

#cross_val_score(scoring="neg_mean_squared_error")
print("5 folds의 개별 neg_mse_scores:",np.round(neg_mse_scores,2))
print("5 folds 개별 rmse scores:",np.round(rmse_scores,2))
print("평균 rmse:{:.3f}".format(avg_rmse))
```

```python
# 릿지에 사용될 alpha 파라미터의 값을 정의
alphas=[0,0.1,1,10,100]

for alpha in alphas:
    ridge=Ridge(alpha=alpha)
    #cross_val_score()로 5폴드세트로 mse 구현후 이를 기반으로 다시 rmse구함 
    neg_mse_scores=cross_val_score(ridge,X_data,y_target,scoring="neg_mean_squared_error",cv=5)
    rmse_scores=np.sqrt(-1*neg_mse_scores)
    avg_rmse=np.mean(rmse_scores)

    #cross_val_score(scoring="neg_mean_squared_error")
    print("alpha{0}일때, 평균 rmse:{1:.3f}".format(alpha,avg_rmse))
```

```python
# 각 alpha에 따른 회귀 계수 값을 시각화하기 위해 5개의 열로 된 맷플롯립 축 생성 
fig,axs=plt.subplots(figsize=[20,6],nrows=1,ncols=5)
coeff_df=pd.DataFrame()
#alphs 리스트 값을 차례로 입력해 회귀 계수 값 시각화 및 데이터 저장. pos는 axis의 위치 선정 
for pos,alpha in enumerate(alphas):
    ridge=Ridge(alpha=alpha)
    ridge.fit(X_data,y_target)
    #alpha에 따른 피처별로 회ㅟ계수를 series로 변환하고 이를 df칼럼으로추가 
    coeff=pd.Series(data=ridge.coef_,index=X_data.columns)
    colname="alpha:"+str(alpha)
    coeff_df[colname]=coeff
    #막대그래프로 각 alpha 값에서의 회귀 계수를 시각화, 회귀 계수 값이 높은 순으로 표현 
    coeff=coeff.sort_values(ascending=False)
    axs[pos].set_title(colname)
    axs[pos].set_xlim(-3,6)
    sns.barplot(x=coeff.values,y=coeff.index,ax=axs[pos])
plt.show()
```

```python
sort_column="alpha:"+str(alphas[0])
coeff_df.sort_values(by=sort_column,ascending=False)
```
