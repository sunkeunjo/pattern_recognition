```python
import os 
import pandas as pd 
import numpy as np
import statsmodels.api as sm
#데이터 호출용 
from sklearn.datasets import load_boston
```

```python
# 현재 경로 확인 
os.getcwd()
```

```python
# 데이터 불러오기 
boston=load_boston()
```

```python
#boston data
df=pd.DataFrame(boston['data'],columns=boston['feature_names'])
df['price']=boston['target']
df.head()
```

```python
# target 제외한 데이터만 뽑기
df.drop(['price'],axis=1,inplace=True)
df.head()
```

```python
# data 통계 뽑아보기 
df.describe()
```

```python
'''
타겟 데이터 
1978 보스턴 주택 가격 506개 타운의 주택가격 중앙값(단위, 1000달러 )

특징 데이터 
CRIM: 범죄율 
ZN : 25000평방피트를 초과 거주지역 비율 
INDUS:비소매상업지역 면적 비율 
NOX : 일산화질소 농도
RM: 주택당 방수 
LSTAT: 인구 중 하위 계층 비율 
B: 인구 중 흑인 비율 
PTRATIO: 학생/교사 비율 
CHAS: 찰스강의 경계에 위치한 경우는 1 아니면 0
AGE: 1940년 이전에 건축된 주택의 비율 
RAD: 방사형 고속도로까지의 거리 
DIS: 직업센터의 거리
TAX: 재산세율

'''
```

# CRIM/RM/LSTAT 세 개의 변수로 각각 단순 선형 회귀 분석하기 


```python
# 변수 설정 TARGET/CRIM/RM/LSTAT
df=df.loc[:,["CRIM",'RM','LSTAT']]
df['target']=boston['target']
df.head()
```

```python
# 각각에 배분하기 
target=df[['target']]
crim=df[['CRIM']]
RM=df[['RM']]
lstat=df[['LSTAT']]
```

# target~crim 선형회귀 분석 

```python
# crim 변수에 상수항 추가하기 
crim1=sm.add_constant(crim,has_constant="add")
crim1
```

```python
# sm.OLS 적합시키기 
model1=sm.OLS(target,crim1)
fitted_model1=model1.fit()
```

```python
# summary함수통해 결과 출력
fitted_model1.summary()
```

```python
# 회귀 계수 출력
fitted_model1.params
```

y_hat=beta0+bet1*x계산해보기 

```python
#회귀 계수 x 데이터 (x)
np.dot(crim1,fitted_model1.params)
```

```python
#  predict 함수를 통해 yhat 구하기 
pred1=fitted_model1.predict(crim1)
```

```python
# 직접구한 yhat과 predict 함수를 통해 구한 yhat 차이 
(np.dot(crim1,fitted_model1.params)-pred1).sum()
```

# 적합시킨 직선 시각화 

```python
import matplotlib.pyplot as plt 
plt.yticks(fontname="Arial")
plt.scatter(crim,target,label="data")
plt.plot(crim,pred1,label="result",c="r")
plt.legend()
plt.show()
```

```python
plt.scatter(target,pred1)
plt.xlabel("real_value")
plt.ylabel("pred_value")
plt.show()
```

# 이걸 통해서도 직선이 아직 맞는 것 같지는 않다는 것을 확인할 수 있다. 

```python
## residual 시각화 
fitted_model1.resid.plot()
plt.xlabel("Residual_number")
plt.show()
```

```python
# 잔차의 합계산하기
np.sum(fitted_model1.resid)
```

# 위와 동일하게 target\~rm / target\~ lstat에 대해서도 적용시켜 보자.

```python
rm1 =sm.add_constant(RM,has_constant="add")
rm1.head()
```

```python
model2=sm.OLS(target,rm1)# 순서를 잘 맞춰줘야합니다.
fitted_model2=model2.fit()
# target에 넣고 model을 만들고 맞춥니다. 
```

```python
#summary 함수를 통한 출력 
fitted_model2.summary()
```

```python
# 회귀꼐수 출력 
fitted_model2.params
```

```python
# y hat값 구하기 -> X*beta 
np.dot(rm1,fitted_model2.params)
```

```python
#predict 값구하기 
pred2= fitted_model2.predict(rm1)
# 예측값과 직접 구한 값 사이에 차이는 없다.
print((pred2-np.dot(rm1,fitted_model2.params)).sum())
```

```python
#적합시킨 직선 시각화 
import matplotlib.pyplot as plt 
fig , ax= plt.subplots(3,1,figsize=[10,6])
ax[0].scatter(RM,target,label='data')
ax[0].plot(RM,pred2,label='predicted',color='r')
ax[0].legend()
ax[1].scatter(target,pred2)
ax[1].set_xlabel("real_vlaue")
ax[1].set_ylabel("pred_value")
ax[2].plot(fitted_model2.resid)
ax[2].set_xlabel("residual_number")
plt.show()
```

```python
# lstat 시작 
lstat1=sm.add_constant(lstat,has_constant="add")
lstat1.head()
```

```python
# 회귀모델에 적합 
model3 = sm.OLS(target,lstat1)
fitted_model3=model3.fit()
```

```python
# model summary 
fitted_model3.summary()
```

```python
# predict 값과 hat값의 차이가 있는지 확인 
pred3= fitted_model3.predict(lstat1)
print((np.dot(lstat1,fitted_model3.params)-pred3).sum())
```

```python
fitted_model3.resid
```

```python
fitted_model3.resid.cumsum()/len(fitted_model3.resid)
```

```python
# 시각화 
fig, ax =plt.subplots(3,1,figsize=[20,10])
ax[0].scatter(lstat,target,label="data")
ax[0].plot(lstat,pred3,label="predicted",color='r')
ax[0].legend()
ax[1].scatter(target,pred3)
ax[1].set_xlabel("real_value")
ax[1].set_ylabel("predicted_value")
ax[2].plot(fitted_model3.resid)
plt.show()
```
