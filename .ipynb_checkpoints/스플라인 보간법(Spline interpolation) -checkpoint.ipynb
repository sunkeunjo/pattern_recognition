# 스플라인 보간법(Spline interpolation) 

___

스플라인 보간법은 다항 함수의 기저함수에서 갖는 문제점인 "입력 공간의 한 영역에서 발생한 변화가 다른 영역들에게 까지 영향을 끼친다"를 보완하기 위해서 나온 방법이다.

__wikipedia__ 에 나오는 정의에 의하면,<br>
> 전체 구간을 소구간 별로 나누어서 저차수의 다항식으로 매끄러운 함수를 구하는 방법이라고 나온다

스플라인 보간법의 특징은 
> * 전체에서 어느 한 부분만 급격히 변하는 함수의 움직임에 우수한 근사를 제공해준다. 
* 낮은 차수의 다항식에서 사용하기에 유의미하다

이러한 스플라인 보간법을 적용하기 위해서는 다음과 같은 조건이 요구된다. 
> * n개 데이터점, (n-1)개 소구간, 각 소구간 i, 소구간별 스플라인 함수 $s_j$가 주어질 때, 
    - 각 소구간에서 보간점이 정의될 수 있어야 된다.($y_i=s_i(x_i)(i=0,1,2,...m)$
    - 각 소구간에서 n-1차 연속 미분이 가능해야 한다
    - 각 소구간에서 n차 다항식으로 표현이 가능해야한다

# 수학적으로 분석하기

어떤 곡선이 `부드럽다`는 것의 의미는 연속함수임과 동시에 곡선의 변화 역시 부드럽다는 뜻입니다. 수학적으로 "변화"라는 것은 미분으로 우리는 해석할 수 있기 때문에, 어떤 곡선이 부드러우려면 1계 미분도 연속이어야 함을 알 수 있습니다. 더 나아가서, 곡선의 __곡률__도 부드러웠으면 좋을 것 같습니다.  `곡률`이라는 것은 곡선이 굽은 정도인데, 곡선의 각 점마다 굽은 정도가 다 다를 것입니다. 그런데 굽어짐이라는 것도 적당히 부드럽다면, 곡선이 전체적으로 부드러워보일 것입니다. 곡률은 수학적으로 2계 미분, 즉 두 번 미분한 것에 대응하므로 , 2계 미분도 연속이어야 함을 알 수 있습니다. 이를 정리하면, 우리가 원하는 부드러운 곡선은 $f \in C^2$, 즉 2번 미분하여 연속인 함수여야 합니다.

# 공식 세우기 

위에서 세운 원리를 적용하여 공식을 세워보겠습니다. 우리는 보간하는 곡선이 __1. 두 번 미분하여 연속이고 2. 계산이 간단하기를__ 기대합니다. 2번 조건인 간단한 계산은 다항식이면 충분하므로, 우리는 결국 두 번 미분하여 연속인 다항식으로 보간합니다. 이런 다항식 중 차수가 가장 낮은 다항식은 결국 3차 함수 입니다. 그래서 이 알고리즘은 보통 __Cubic Spline Interpolation__이라고 합니다. 

먼저 보간할 점을 $x_0,x_1,x_2,...x_N$이라고 하겠습니다. 그리고 각 $x_k$에 대응하는 y축 상의 점은 $f_k$라고 하겠습니다. 즉, 우리는 $(x_0,f_0),...(x_N,f_N)$인 점들을 대상으로 부드럽게 잇습니다.

즉, 각 데이터 $x_i,x_{i+1}$은 곡선 $S_i$로 연결을 시키는데, 각 $S_i$가 3차 다항식이라고 가정하는 것이 Cubic Spline interpolation이다.
이후의 편의를 위해, $$\begin{align}
A(x) &:=\frac{x_{k+1}-x}{x_{k+1}-x_{k}} \\
B(x) &:= 1 - A = \frac{x - x_{k}}{x_{k+1}-x_{k}} \\
C(x) &:=\frac{1}{6} (A^3 - A) (x_{k+1}-x_{k})^2 \\
D(x) &:=\frac{1}{6} (B^3 - B) (x_{k+1}-x_{k})^2 \\
\end{align}$$
로 정의하겠습니다. 그리고 $(x_k,f_k)$에서의 2계 미분 값은 $f''_k$로 쓰겠습니다. 이렇게 쓰면, 우리는 $ x_k,x_{k+1} $ 사이에서 보간할 함수는 $ F_{k}(x) = A(x) f_k + B(x) f_{k+1} + C(x) f_{k}^{''} + D(x) f_{k+1}^{''} $ 가 된다는 것을 알 수 있습니다. $F_k(x) $ 가 저렇게 생긴 이유는 $$\begin{align}
F_{k}(x_k) &= f_k \\
F_{k}(x_{k+1}) &= f_{k+1} \\
F_{k}^{‘’} (x_k) &= f_{k}^{‘’} \\
F_{k}{‘’} (x_{k+1}) &= f_{k+1}^{‘’} \\
\end{align}$$

이기 떄문이고 위의 식을 만족하는 3차함수는 A,B,C,D 밖에 없기 때문입니다! 

이제 위의 식들에 $f_{0}^{''} = f_{N}^{''} = 0$을  조건으로 주어 방정식을 구하게 되면 우리는 보간된 식을 구할 수 있습니다.

```python
import scipy.interpolate as ip
from scipy.interpolate import splrep,splev
import matplotlib.pyplot as plt 

print("x값은 ",x)
# 원래 y의 값들을 넣어봅시다. 
y=np.cos(x)
print(y)
```

```python
# x,y값을 주고 추정하는 스플라인 곡선을 만듭니다.
spl=splrep(x,y)
print(spl[0]) # 은 1x 19의 값을 가지고 있습니다. 원래의 x값보다 4개가 더 생겼습니다.
```

```python
#0~10까지의 50 구간에 대한 모든 점들을 위 스플라인 곡선으로 추정한 y1값을 구해봅시다
x1=np.linspace(0,10,50)
y1=splev(x1,spl)# x1식에 spl을 집어넣어서 보는 것이다

#그림을 그려 봅니다
plt.figure(figsize=(16,5))
plt.plot(x,y,"ro-")
plt.plot(x1,y1,"b")
plt.legend(['original','splined function'])
plt.title("basis function is cosine")
plt.grid()
plt.show()
```

```python
#이번에는 sine 곡선으로 그려봅시다
print(x)
y=np.sin(x)
spl=splrep(x,y)

#cos과 마찬가지로 x값에 더 넓은 범위의 값을 집어넣는다 
x1=np.linspace(0,10,50)
y2=splev(x1,spl)# x1을 spl에 넣어서 본다 

#다시 그려봅니다 
plt.figure(figsize=(16,5))
plt.plot(x,y,'ro-')
plt.plot(x1,y2,'b')
plt.legend(['original','splined function'])
plt.title("basis function is sine")
plt.grid()
plt.show()
```

```python
# 전체 코드 
import numpy as np
import scipy.interpolate as ip 
from scipy.interpolate import splrep, splev
import matplotlib.pyplot as plt 

# 0~10 까지 15개로 나누어 점을 찍음 
x0=np.linspace(0,10,15)
print("x0= ",x0)
#cosine 값을 계산 
y0=np.cos(x0)


# x,y 값을 주고 추정하는 스플라인 곡선을 만든다
spl=splrep(x0,y0)
# 0~10까지 50구간에 대한 모든 점들을 위 스플라인 곡선으로 추정한 y값을 구한다. 
x1=np.linspace(0,10,50)
y1=splev(x1,spl)

#그린다
plt.figure(figsize=(16,5))
plt.subplot(121)
plt.plot(x0,y0,"-")
plt.plot(x1,y1,"r")
plt.grid()

#이번에는 sine곡선으로 추정해 본다.
plt.subplot(122)
y2=np.sin(x0)
spl2=splrep(x0,y2)
y3=splev(x1,spl2)
plt.plot(x0,y2,"-")
plt.plot(x1,y3,'r')
plt.grid()
plt.show()
```
