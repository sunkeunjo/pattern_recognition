# AFINN-111 Sentiment Analysis 

Text sentiment analysis(or usually just sentiment analysis) is a text mining technique to analyze the sentiment of the writer or to the topic written about. One approach is to use a word list where each word has been scored for positively/negatively or sentiment strength. There are different lists, we use AFFIN and the version 111. 

> AFINN is a list of English words rated for valence with an integer between minus five (negative) and plus five (positive). The words have been manually labeled by Finn Årup Nielsen in 2009-2011. The file is tab-separated. There are two versions:
<br>
AFINN-111: Newest version with 2477 words and phrases.

```python
#Get requests response "Failed to establish a new connection: [Errno -2]"
url_web = "https://raw.githubusercontent.com/hyharryhuang/AFINN-json/master/AFINN.json"
#We are going to hardcode the Afinn-111 json
```

```python
afinn_data=rq.get(url_web)
afinn_data.status_code
```

```python
# string into dictionary type
import ast
# scre_word_dict = dict(afinn_data)
score_word_dict=ast.literal_eval(afinn_data.text)
```

and lets set the score of each word definition 

```python
# set the score of each word definition 
def score_definition(x,score_word_dict):
    score=0
    words=x.lower().split(" ")
    for word in words:
        if (word in score_word_dict):
            score+=int(score_word_dict[word])
    return score
```

```python
urban_data['score']=urban_data['definition'].apply(lambda x: score_definition(x,score_word_dict))
```

Words with the worst sentiment 

```python
urban_data.sort_values(by="score",ascending=True)
```

```python
worst_sentiment_words=urban_data.sort_values(by="score",ascending=True)['word'][:25].to_string().replace("$","s")
wordcloud=WordCloud(background_color="white",relative_scaling=0,normalize_plurals=True).generate(worst_sentiment_words.lower())

plt.figure(figsize=[15,10])
plt.title("Worst Sentiment Words")

plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```
