# Activation function(활성화 함수란?)
> 입력신호의 총합을 출력신호로 변환하는 함수를 일반적으로 활성화 함수라고 합니다. 활성화 함수는 입력 신호의 총합이 활성화를 일으키는지 정하는 역할를 합니다. 
<center> 출처: Deep learning from scratch </center>

# 활성화 함수의 종류 3가지 

## 1. 계단 함수(step function) 
>  입력값을 0 또는 1의 값으로 출력하는 함수 

## 2. 시그모이드 함수(sigmoid function)
> 입력값을 0~1 사이의 실수로 출력하는 함수 

## 3. ReLu(Rectified Linear Units) 
> 입력이 0이 넘으면 그 입력값을 그대로 출력하고 0이하이면 0을 출력하는 함수

```python
# 퍼셉트론에서는 활성화 함수로 계단 함수를 이용한다. 
import numpy as np 
def step_function(x_data):
    result=[0 if x<=0 else 1 for x in x_data]
    
    return result 
```

```python
x_data=np.arange(-10,10,0.1)

import matplotlib.pyplot as plt 
%matplotlib inline 

plt.plot(x_data,step_function(x_data))
plt.title("Step Function")
plt.show()
```

# 시그모이드 함수 
___

<center> ※ 계단함수와 시그모이드 함수의 공통점 & 차이점</center>

공통점]
> 두 함수 모두 0과 1 사이의 데이터만 출력한다. 비선형 함수이다. 

차이점] 
> 계단함수는 숫자 1과 0만 출력하는데 반해, 시그모이드 함수는 0과 1사이의 실수를 출력한다. 

[시그모이드 함수식]

$$h(x)=\frac{1}{1+exp(-x)}$$

"통계학에서 성공할 확률이 실패할 확률보다 얼마나 큰지를 나타내는 오즈비율이라는 값이 있다!"

$$ODDS RATIO = 성공/실패 = P/(1-P)$$

성공할 확률 P를 0에서 1사이의 값으로 나타내면, 실패할 확률은 1-P이다. <그래프1>

밑에 그래프에서 확인할 수 있듯이 P(성공할 확률)가 1에 가까워지면 가까워질 수록 오즈 비율의 값이 급격히 커지는 현상이 발생한다. 

그러한 현상을 방지하기 위해, 우리는 로그를 취한 Logit이라는 함수를 만들어서 작은 숫자로 표현을 해보겠다.<그래프2>

$$Logit Function = log(P/(1-P))=Wx+b$$

이 로짓 함수를 뉴런의 출력값에 따라 확률 p를 구하기 쉽도록 지수함수 형태로 바꾸면 위에서 확인한 시그모이드 함수가 됩니다.

```python
# odds ratio 그래프 
p=np.arange(0,1,0.001)
plt.plot(p,p/(1-p))
plt.title("graph1")
plt.ylabel("odds ratio")
plt.show()
```

```python
# logit function 
p=np.arange(0,1,0.001)

def logit_function(x):
    return np.log(x/(1-x))

y=logit_function(p)
plt.plot(p,y)
plt.title("graph2")
plt.show()
```

```python
# sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.arange(-10,10,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.title("sigmoid function")
plt.show()
```

# 3. ReLu 함수(Rectified Linear Units function) 

"최근에는 활성화 함수로 ReLU함수를 주로 이용합니다" 



```python
x=np.arange(-10,10,0.1)

def relu(x_data):
    return [x if x>=0 else 0 for x in x_data]

y=relu(x)
plt.plot(x,y)
plt.title("ReLU function")
plt.show()
```
