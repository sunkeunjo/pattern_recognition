Given the two sets of integers s1 and s2, the task is to find the Jaccard Index and the Jaccard Distance btw the two sets

The Jaccard Index and the Jaccard Distance between the two sets can be calculated by using the formula



```python
import numpy as np 

def intersection(s1,s2):
    intersection=s1&s2
    return intersection

def Union(s1,s2):
    return s1|s2

def Jaccard_(s1,s2):
    ## input : s1,s2 두개의 set 형식이 들어가고 
    ## output :  Jacard index와 Jaccard distance를 뽑아줍니다. 
    
    index = np.round(len(intersection(s1,s2))/len(Union(s1,s2)),3)
    distance =1-index 
  
    return index,distance

if __name__=="__main__":
    s1=set([1,2,3,4,5])
    s2=set([4,5,6,7,8,9,10])
#     print(intersection(s1,s2))
#     print(Union(s1,s2))
    Jaccard= Jaccard_(s1,s2)
    print(f"Jaccard index= {Jaccard[0]}")
    print(f"Jaccard distance= {Jaccard[1]}")
    s1=set([1,2,3,4,5])
    s2=set([4,5,6,7,8])
    Jaccard=Jaccard_(s1,s2)
    print(f"Jaccard index= {Jaccard[0]}")
    print(f"Jaccard distance= {Jaccard[1]}")  
```
