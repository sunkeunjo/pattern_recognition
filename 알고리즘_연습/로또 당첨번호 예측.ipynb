```python
# 필요한 패키지 
import pandas as pd 
import requests 
from tqdm import tqdm 
import json
from time import sleep
```

```python
# 데이터 베스 만들기 
def getLottoWinInfo(mindrwno, maxdrwno):
       # 이 방법은 원격 호스트에 의해 강제적으로 끊기게 된다. 이를 방지하기 위해서 3초간의 delay를 의도적으로 줘보겠습니다.

    drw1=[];drw2=[];drw3=[];drw4=[];drw5=[];drw6=[];bnus=[]
    totsellamnt=[];date=[];firstaccumamnt=[];firstprize=[];firstwindamnt=[]
    for i in tqdm(range(mindrwno,maxdrwno+1,1)):
        req_url="http://www.nlotto.co.kr/common.do?method=getLottoNumber&drwNo=" + str(i)
        sleep(3)
        req_lotto=requests.get(req_url)
        
        lottono=req_lotto.json()
        drw1.append(lottono["drwtNo1"])
        drw2.append(lottono['drwtNo2'])
        drw3.append(lottono['drwtNo3'])
        drw4.append(lottono['drwtNo4'])
        drw5.append(lottono['drwtNo5'])
        drw6.append(lottono['drwtNo6'])
        bnus.append(lottono['bnusNo'])
        totsellamnt.append(lottono['totSellamnt'])
        date.append(lottono['drwNoDate'])
        firstaccumamnt.append(lottono['firstAccumamnt'])
        firstprize.append(lottono['firstPrzwnerCo'])
        firstwindamnt.append(lottono['firstWinamnt'])
        
        lotto_dict={"추첨일":date, "Num1":drw1, "Num2":drw2, "Num3":drw3, "Num4":drw4, "Num5":drw5, "Num6":drw6, "bnsNum":bnus, "총판매금액":totsellamnt, "총1등당첨금":firstaccumamnt, "1등당첨인원":firstprize, "1등수령액":firstwindamnt}
    df=pd.DataFrame(lotto_dict)
    return df 
```

```python
lotto_df=getLottoWinInfo(1,907)
```
